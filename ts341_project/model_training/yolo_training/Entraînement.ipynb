```python
import yaml
import glob
import random
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(
    style="darkgrid", font_scale=1.5, font="SimHei", rc={"axes.unicode_minus": False}
)

import torch
from ultralytics import YOLO

from PIL import Image

seed = 1
random.seed(seed)
np.random.seed(seed)
torch.manual_seed(seed)
torch.cuda.manual_seed(seed)
```

# <div style="border-radius:10px; border:#E75480 solid; padding: 15px; background-color: #FFFAF0; font-size:100%; text-align:center">Modeling</div>

```python
model = YOLO("yolov8n.pt")
```

```python
config = {
    "path": "ts341_project/model_training/sim2real_approach/final_color_dataset",
    "train": "train",
    "val": "valid",
    "nc": 1,
    "names": ["drone"],
}

with open("data.yaml", "w") as file:
    yaml.dump(config, file, default_flow_style=False)
```

```python
!wandb disabled
```

```python
results = model.train(data="data.yaml", epochs=100, save_period=10, seed=seed, device=0)
```

```python
metrics = model.val()
```

```python
print(metrics.box.map)
```

# <div style="border-radius:10px; border:#E75480 solid; padding: 15px; background-color: #FFFAF0; font-size:100%; text-align:center">Predict Test Data</div>

```python
paths = glob.glob(
    "/kaggle/input/yolo-drone-detection-dataset/drone_dataset/valid/images/*"
)
```

```python
n = 8
results = model.predict(paths[:n])
for i in range(n):
    r = results[i]
    img = Image.fromarray(r.plot())
    plt.figure(dpi=100)
    plt.imshow(img)
    plt.axis("off")
    plt.tight_layout()
    plt.show()
```
